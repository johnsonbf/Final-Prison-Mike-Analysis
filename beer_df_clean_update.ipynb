```python
# Import dependencies
import pandas as pd
```

# Read in Raw Data

```python
# Import csv data into a dataframe
beer_df = pd.read_csv("../Resources/beer_reviews.csv")
beer_df.head()
```

# Drop Null Values 

```python
#Checking for null values
beer_df.isnull().sum()
```

```python
#Dropping null values
beer_df.dropna(inplace=True)
```

```python
beer_df.head()
```

# Unix to Date + Time

```python
# Convert the unix timestamp to datetime
beer_df['review_time'] = pd.to_datetime(beer_df['review_time'], unit='s')
```

```python
# Format the datetime as a string
beer_df['review_time'] = beer_df['review_time'].dt.strftime('%m/%d/%Y %I:%M%p')
```

```python
beer_df.head()
```

# Rename Headers

```python
beer_df = beer_df.rename(columns={'brewery_id': 'Brewery ID', 'review_time': 'Review Time', 'review_overall': 'Overal Review', 'review_aroma': 'Aroma Review', 'review_appearance': 'Appearance Review', 'review_profilename': 'Reviewer Name', 'beer_style': 'Beer Style', 'review palate': 'Palate Review', 'review_taste': 'Taste Review', 'beer_name': 'Beer Name', 'beer_abv': 'ABV', 'beer_beerid': 'Beer ID', 
                      'brewery_name': 'Brewery Name'})
```

```python
beer_df.columns
```

```python
beer_df.head()
```

# Export to CSV

```python
beer_df.to_csv('clean_beer.csv')
```
